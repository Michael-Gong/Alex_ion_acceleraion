```python
%matplotlib inline
#import sdf
import matplotlib
#matplotlib.use('agg')
import matplotlib.pyplot as plt
import numpy as np
from numpy import ma
from matplotlib import colors, ticker, cm
from matplotlib.mlab import bivariate_normal
from optparse import OptionParser
import os
from mpl_toolkits.mplot3d import Axes3D
import random
from mpl_toolkits import mplot3d
from matplotlib import rc
import matplotlib.colors as mcolors


font = {'family' : 'monospace',
        'color'  : 'black',
        'weight' : 'normal',
        'size'   : 20
        }

def beta_to_gamma(beta):
    return 1.0/(1-beta**2)**0.5

def beta_to_Ek(beta):
    return (1.0/(1-beta**2)**0.5-1.0)*1836.*0.51

def gamma_to_beta(gamma):
    return (1-1.0/gamma**2)**0.5

def Ek_to_beta(Ek):
    gamma = Ek/(1836*0.51)+1.0
    return gamma_to_beta(gamma)

if __name__ == '__main__':
    beta_g = np.linspace(0.01,0.99,2000)
    pot_e  = np.linspace(1,400,4000) # potential well

    beta_g,pot_e =  np.meshgrid(beta_g,pot_e,indexing='ij')

    En_1   = beta_to_Ek(beta_g)
    beta_1 = Ek_to_beta(En_1-pot_e) 
    beta_2 = (-beta_1+beta_g)/(1-beta_1*beta_g)
    En_f0  = beta_to_Ek(beta_2)

    beta_f = 2*beta_g/(1+beta_g**2)
    En_f   = beta_to_Ek(beta_f)
    En_f[ En_1 >= pot_e ] =  En_f0[ En_1 >= pot_e ]
#    En_f[ En_1 < pot_e ]  = 0 #  En_f0[ En_1 < pot_e ]

    plt.subplot(2,1,1)
    eee = np.max(En_f)
    levels = np.logspace(0, np.log10(eee), 40)
    plt.contourf(beta_g.T, pot_e.T, En_f.T, levels=levels, norm=mcolors.LogNorm(vmin=1, vmax=eee), cmap=cm.jet)
    #### manifesting colorbar, changing label and axis properties ####
    #cbar=plt.colorbar(ticks=[0, eee/2, eee])
    cbar=plt.colorbar(ticks=[1,10,100,1000,10000])

    cbar.set_label('$E_k$ [MeV]',fontdict=font)
    cbar.ax.set_yticklabels(cbar.ax.get_yticklabels(),fontsize=15)
    plt.xlabel(r'$\beta_g$',fontdict=font)
    plt.ylabel(r'$\psi$ [MeV]',fontdict=font)
    plt.xticks(fontsize=20); plt.yticks(fontsize=20);
    plt.title('Ion acceleration',fontdict=font)

    plt.subplot(2,1,2)
    plt.plot(beta_g[:,498],  En_f[:,498],  '-k', linewidth=4, label=r'$\psi$ = 50 MeV')
    plt.plot(beta_g[:,1498], En_f[:,1498], '-g', linewidth=4, label=r'$\psi$ = 150 MeV')
    plt.plot(beta_g[:,2498], En_f[:,2498], '-r', linewidth=4, label=r'$\psi$ = 250 MeV')
    plt.plot(beta_g[:,3498], En_f[:,3498], '-b', linewidth=4, label=r'$\psi$ = 350 MeV')

    plt.xlabel(r'$\beta_g$',fontdict=font)
    plt.ylabel('$E_k$ [MeV]',fontdict=font)
    plt.xticks(fontsize=20); 
    plt.yticks(fontsize=20);
    plt.grid(which='major',color='k', linestyle='--', linewidth=0.3)
    plt.grid(which='minor',color='k', linestyle='--', linewidth=0.1)

    plt.legend(loc='best',fontsize=18,framealpha=1.0)


    fig = plt.gcf()
    fig.set_size_inches(10, 17)
    #fig.set_size_inches(5, 4.5)
    #fig.savefig('./summary/a=0_b=0.01_py=100.png',format='png',dpi=1280)
    #plt.close("all")
```
